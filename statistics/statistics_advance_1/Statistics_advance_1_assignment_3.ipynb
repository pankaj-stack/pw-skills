# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Hypothesis testing is a fundamental statistical method used to make decisions or draw conclusions about a population based on sample data. It involves setting up two mutually exclusive hypotheses: the null hypothesis (H0) and the alternative hypothesis (Ha). The purpose of hypothesis testing is to determine whether there is enough evidence in the sample data to reject the null hypothesis in favor of the alternative hypothesis.

Here's a breakdown of the key components of hypothesis testing:

1. Null Hypothesis (H0): This is the default or initial assumption that there is no significant effect, difference, or relationship in the population. It represents the status quo or no change.

2. Alternative Hypothesis (Ha): This is the statement that contradicts the null hypothesis and suggests there is a significant effect, difference, or relationship in the population.

3. Test Statistic: A mathematical formula or procedure is used to calculate a test statistic from the sample data. The choice of test statistic depends on the nature of the data and the hypothesis being tested.

4. Significance Level (α): This is the predetermined level of significance that sets the threshold for how much evidence is required to reject the null hypothesis. Commonly used values include 0.05 (5%) and 0.01 (1%).

5. P-value: The p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis.

6. Decision: Based on the p-value and the chosen significance level, a decision is made. If the p-value is less than or equal to the significance level, the null hypothesis is rejected. If the p-value is greater than the significance level, the null hypothesis is not rejected.

7. Conclusion: The conclusion of the hypothesis test is stated in terms of the null and alternative hypotheses. For example, if the null hypothesis is rejected, it may be concluded that there is a statistically significant effect or difference.

Importance of Hypothesis Testing:

1. **Informed Decision-Making**: Hypothesis testing provides a systematic and objective way to make decisions based on data. It helps researchers and decision-makers assess the impact of interventions, treatments, or changes in processes.

2. **Scientific Validity**: It is a critical tool in scientific research, allowing researchers to determine whether their findings are statistically significant or merely due to chance.

3. **Quality Control**: In industries and manufacturing, hypothesis testing is used to ensure the quality of products and processes by identifying deviations from established standards.

4. **Risk Assessment**: It helps assess the risk associated with making decisions. By setting a significance level, decision-makers can control the risk of making a Type I error (incorrectly rejecting a true null hypothesis).

5. **Policy and Strategy Development**: In fields like economics and social sciences, hypothesis testing is used to assess the effectiveness of policies, strategies, and programs.

6. **Medical Research**: In medical and pharmaceutical research, hypothesis testing is crucial for determining the efficacy of treatments, drugs, and medical procedures.

7. **Continuous Improvement**: In business and management, hypothesis testing is used to identify areas for improvement and to make data-driven decisions for organizational growth.

In summary, hypothesis testing is a foundational statistical method that enables data-driven decision-making, helps validate research findings, and plays a critical role in various fields by providing a structured approach to drawing conclusions from data.

# Q6: What is a null and alternative hypothesis? Give some examples.


In hypothesis testing, the null hypothesis (H0) and the alternative hypothesis (Ha) are two complementary statements that help frame a statistical test and guide the decision-making process. These hypotheses are used to assess whether there is evidence to support a particular claim or hypothesis about a population parameter, such as a mean, proportion, or variance. Here's an explanation and examples of both:

1. **Null Hypothesis (H0)**:
   - The null hypothesis is a statement of no effect, no difference, or no relationship. It represents the status quo or the default assumption.
   - It is typically denoted as H0 and is a statement that the population parameter is equal to a specific value or follows a particular pattern.
   - The null hypothesis is what we aim to test against.

   Examples:
   - H0: The average height of adult males in a particular region is 175 centimeters.
   - H0: The proportion of customers who prefer product A is 0.40.
   - H0: There is no difference in test scores before and after a teaching intervention (μ_before = μ_after).

2. **Alternative Hypothesis (Ha)**:
   - The alternative hypothesis is a statement that contradicts the null hypothesis. It represents the claim or hypothesis you want to provide evidence for.
   - It is typically denoted as Ha or sometimes as H1.
   - The alternative hypothesis can take several forms, depending on what you are trying to show. Common forms include greater than (>) or less than (<), indicating a one-tailed test, or not equal to (!=), indicating a two-tailed test.

   Examples:
   - Ha: The average height of adult males in a particular region is greater than 175 centimeters.
   - Ha: The proportion of customers who prefer product A is less than 0.40.
   - Ha: There is a difference in test scores before and after a teaching intervention (μ_before ≠ μ_after).

In hypothesis testing, you collect sample data and use statistical tests to determine whether the evidence from the sample supports the null hypothesis or provides enough evidence to reject it in favor of the alternative hypothesis. The choice of null and alternative hypotheses depends on the research question and what you want to investigate or prove. The p-value, significance level, and the test statistic are used to make this determination.

Hypothesis testing is a structured process used in statistics to make decisions or draw conclusions about a population based on sample data. The following are the general steps involved in hypothesis testing:

1. **Formulate Hypotheses**:
   - **Null Hypothesis (H0)**: Start by stating the null hypothesis, which represents the default assumption or no effect, no difference, or no relationship in the population.
   - **Alternative Hypothesis (Ha)**: State the alternative hypothesis, which is the claim or hypothesis you want to provide evidence for. It contradicts the null hypothesis.

2. **Set Significance Level (α)**:
   - Choose a significance level (α), often denoted as 0.05 (5%) or 0.01 (1%). This represents the maximum acceptable probability of making a Type I error (incorrectly rejecting a true null hypothesis).

3. **Collect and Prepare Data**:
   - Gather sample data from the population of interest. Ensure the sample is random and representative.

4. **Choose a Statistical Test**:
   - Select an appropriate statistical test based on the data type and research question (e.g., t-test, chi-square test, ANOVA, correlation, regression).

5. **Calculate Test Statistic**:
   - Use the chosen statistical test to calculate a test statistic from the sample data. This test statistic quantifies the difference, effect, or relationship being tested.

6. **Determine the P-value**:
   - Calculate the p-value, which is the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis.

7. **Compare P-value to Significance Level**:
   - Compare the calculated p-value to the predetermined significance level (α).
   - If p-value ≤ α: Reject the null hypothesis. There is enough evidence to conclude that the alternative hypothesis is likely true.
   - If p-value > α: Fail to reject the null hypothesis. There is not enough evidence to conclude that the alternative hypothesis is true.

8. **Draw a Conclusion**:
   - State the conclusion of the hypothesis test in the context of the research question.
   - If you reject the null hypothesis, interpret the result in favor of the alternative hypothesis.
   - If you fail to reject the null hypothesis, acknowledge that there is insufficient evidence to support the alternative hypothesis.

9. **Report Results**:
   - Communicate the results of the hypothesis test, including the test statistic, p-value, conclusion, and any relevant statistics or effect sizes.

10. **Consider Practical Significance**:
    - While statistical significance is important, consider whether the observed effect or difference is practically significant and has real-world implications.

11. **Repeat if Necessary**:
    - In some cases, it may be necessary to repeat the test with different data or under different conditions to further validate the results.

These steps provide a structured framework for conducting hypothesis testing and making informed decisions based on statistical evidence. The choice of specific tests and details of the procedure may vary depending on the research question and the type of data being analyzed.

# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,given a sample from each population.

```python
import numpy as np
from statsmodels.stats.weightstats import CompareMeans
import seaborn as sns 

#. **Collect Data**:

titanic_df = sns.load_dataset('titanic')
population_age_female = titanic_df[titanic_df.sex=='female'].age # this is the population age
population_age_male = titanic_df[titanic_df.sex=='male'].age # this is the population age

sample_size = 25 # sample size is less than 30 so we will perform t-test not z-test 

sample_age_female = population_age_female.sample(sample_size)# sample age for female
sample_age_male = population_age_female.sample(sample_size)# sample age for male

average_of_males = sample_age_male.mean()
average_of_females = sample_age_female.mean()

significance_level = 0.05 # alpha

# 3. **Define Null and Alternative Hypotheses**:

    # average_of_males = average_of_females  <-------- Null Hypothesis(H0)
    # average_of_males != average_of_females <-------  Alternavtive hypothesis(H1 or Ha)





# 4. **Performing the Test**:
#    Use the `CompareMeans` class from StatsModels to perform the test. Here's an example using a t-test for two independent samples:

cm = CompareMeans.from_data(sample_age_male.dropna(), sample_age_female.dropna())
tstat, pvalue, dof = cm.ttest_ind(alternative='two-sided', usevar='pooled')

print(f'tstat : {tstat}',f'pvalue : {pvalue}',f'dof : {dof}',sep='\n')
# 5. **Interpret the Results**:
#    - `tstat` is the t-statistic.
#    - `pvalue` is the p-value.
#    - `dof` is the degrees of freedom.

# 6. **Make a Decision**:
#    Compare the p-value to your chosen significance level (e.g., 0.05). 
# If p-value < significance level, reject the null hypothesis.

if pvalue < significance_level:
    print('Rejecting the Null Hypothesis <---------------')
else:
    print('Failed to reject the Null Hypothesis <---------------')
    
print(f'average_of_males : {average_of_males}')
print(f'average_of_females : {average_of_females}')
```

# Q8. Define p-value and explain its significance in hypothesis testing.

A p-value, or probability value, is a statistical measure that quantifies the likelihood that observed data would have occurred by random chance, assuming a null hypothesis is true. **It is a number between 0 and 1**. Here's the significance of p-values in hypothesis testing:

1. **Determining Significance**: P-values help researchers assess the significance of their findings. A small p-value (typically ≤ 0.05) suggests that the observed results are unlikely to have occurred by chance, indicating evidence against the null hypothesis.

2. **Null Hypothesis Testing**: In hypothesis testing, researchers start with a null hypothesis (H0) that there is no effect or relationship. The p-value is used to decide whether to accept or reject this null hypothesis. If the p-value is small, the null hypothesis is rejected in favor of the alternative hypothesis.

3. **Threshold for Decision**: The p-value provides a threshold for decision-making. If the p-value is less than or equal to a predefined significance level (often 0.05), researchers reject the null hypothesis, suggesting that their findings are statistically significant.

4. **Interpreting Results**: Researchers use p-values to communicate the strength of their findings to others. A smaller p-value indicates stronger evidence against the null hypothesis.

5. **Caution**: However, it's essential to remember that p-values do not provide the effect size or the practical significance of a result. They only indicate statistical significance. Therefore, it's crucial to consider the context and practical implications of the findings alongside p-values.

In summary, p-values are a crucial tool in hypothesis testing, helping researchers assess the strength of evidence against the null hypothesis and make informed decisions about their research outcomes.

# Q11: What is Student’s t distribution? When to use the t-Distribution.

Student's t-distribution, often simply referred to as the t-distribution, is a probability distribution that is similar to the normal (Gaussian) distribution in many ways. It's named after William Sealy Gosset, who initially developed it under the pseudonym "Student" while working at the Guinness Brewery in the early 20th century. The t-distribution is widely used in statistics and hypothesis testing, especially when dealing with small sample sizes.

Key characteristics of the t-distribution include:

1. **Bell-Shaped Curve:** Like the normal distribution, the t-distribution has a bell-shaped curve. However, it has heavier tails compared to the normal distribution. This means that it has more probability in the tails, which is important when dealing with sample statistics.

2. **Shape Parameter:** The shape of the t-distribution is determined by a single parameter called the degrees of freedom (df). The degrees of freedom depend on the sample size and play a crucial role in shaping the distribution. As the degrees of freedom increase, the t-distribution approaches the shape of the normal distribution.

3. **Variance:** The t-distribution has a variance greater than 1, which contributes to its wider tails. This makes it suitable for dealing with situations where the population standard deviation is unknown and needs to be estimated from the sample.

When to use the t-distribution:

1. **Small Sample Sizes:** The t-distribution is particularly useful when dealing with small sample sizes (typically n < 30). In such cases, the normal distribution assumption may not be valid due to the limited data, and the t-distribution provides more accurate confidence intervals and hypothesis tests.

2. **Unknown Population Standard Deviation:** When the population standard deviation is unknown and needs to be estimated from the sample, the t-distribution is used. In contrast, the normal distribution assumes a known population standard deviation.

3. **Hypothesis Testing:** The t-distribution is commonly used in hypothesis testing scenarios, especially for comparing sample means from two groups (independent samples) or for testing the mean of a single sample against a known or hypothesized value.

4. **Confidence Intervals:** When constructing confidence intervals for population parameters, such as the mean, using sample data, the t-distribution is used to account for the uncertainty associated with small sample sizes.

In summary, the t-distribution is a probability distribution that is particularly valuable when working with small sample sizes and situations where the population standard deviation is unknown. It is commonly used in hypothesis testing and constructing confidence intervals, helping statisticians make inferences about population parameters based on sample data with a greater degree of accuracy.

# Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic (or t-score) is a statistical measure that quantifies how many standard errors a sample mean is away from the population mean. It is used primarily in hypothesis testing and confidence interval construction when dealing with small sample sizes and situations where the population standard deviation is unknown.

The formula for the t-statistic is as follows:

\[t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}\]

Where:
- \(t\) is the t-statistic.
- \(\bar{X}\) is the sample mean.
- \(\mu\) is the population mean (the hypothesized or known value being tested against).
- \(s\) is the sample standard deviation.
- \(n\) is the sample size.

In this formula, the numerator (\(\bar{X} - \mu\)) represents the difference between the sample mean and the population mean, while the denominator (\(\frac{s}{\sqrt{n}}\)) standardizes this difference by dividing it by the standard error of the sample mean. The standard error of the sample mean is a measure of how much sample means are expected to vary from one sample to another. It depends on the sample standard deviation (\(s\)) and the square root of the sample size (\(\sqrt{n}\)).

The t-statistic is then compared to critical values from the t-distribution to determine whether the observed difference between the sample mean and the population mean is statistically significant. If the t-statistic is large enough (far from zero), it suggests that there is evidence to reject the null hypothesis (typically stating that the population mean is equal to a hypothesized value). In other words, a large t-statistic indicates that the sample mean is significantly different from the hypothesized population mean.

The critical values from the t-distribution depend on the chosen significance level (\(\alpha\)) and the degrees of freedom (\(df\)), which are related to the sample size. Researchers typically compare the absolute value of the t-statistic to the critical t-value to make a decision about the null hypothesis. If the t-statistic exceeds the critical t-value, the null hypothesis is rejected; otherwise, it is not.
